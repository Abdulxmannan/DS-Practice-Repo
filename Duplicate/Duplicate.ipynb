## Example 2(Keep First, Last, False)

```python
df_laptops=df_laptops.sort_values(['Company', 'Price_euros'])
df_laptops
```

```python
df_laptops.value_counts('Company')
```

```python
duplicate_first= df_laptops.duplicated('Company', keep='first')
```

```python
df_laptops[duplicate_first]
```

```python
df_laptops[duplicate_first]
```

```python
df_laptops[~duplicate_first] [['Company', 'Price_euros']]
# cheapest laptop according to each company
```

```python
duplicate_Last= df_laptops.duplicated('Company', keep='last')
```

```python
df_laptops[duplicate_Last]
```

```python
df_laptops[~duplicate_Last][['Company', 'Price_euros']]
```
